# Standard Deviation (SD)

**Standard Deviation** measures how spread out numbers in a dataset are from the average (mean). It essentially tells you how "concentrated" the data is.

* **Low SD:** Data points cluster tightly near the mean (indicates stability/consistency).
* **High SD:** Data points are spread widely from the mean (indicates volatility/unpredictability).

> **Note:** SD is the square root of the variance. This returns the metric to the **original units** of the data (e.g., dollars, meters), making it much easier to interpret than variance (which is in squared units).

## Key Concepts

* **Low SD Example:** Consistent test scores (e.g., 88, 90, 92).
* **High SD Example:** Fluctuating stock prices (e.g., $50, $120, $30).

* **Units:** If your data is in *meters*, Variance is in $m^2$, but Standard Deviation returns to *meters*.

## Common Applications

1.  **Risk Assessment:** Used in finance to quantify investment volatility (risk).
2.  **Quality Control:** Used by businesses to monitor product consistency.
3.  **Data Analysis:** Helps determine if data is clustered around a central value or dispersed.

## How to Calculate It

The calculation differs slightly depending on whether you have a whole population or just a sample.

**The Formula (Sample SD):**
$$s = \sqrt{\frac{\sum (x_i - \bar{x})^2}{n - 1}}$$

**Step-by-Step:**
1.  **Find the Mean ($\bar{x}$):** Calculate the average of your data set.
2.  **Calculate Deviations:** Subtract the mean from each individual data point $(x_i - \bar{x})$.
3.  **Square Deviations:** Square each result to remove negative values.
4.  **Sum Squares:** Add up all the squared deviations.
5.  **Calculate Variance:** Divide by $n-1$ (for a sample) or $N$ (for a population).
6.  **Find SD:** Take the square root of the variance to get the final Standard Deviation.
