```python
# rows
df.loc[0]
df.loc[0:2] # start idx : ending idx (inclussive) 

df.iloc[2]
df.iloc[0:2] # start idx : ending idx (exclusive)

# cells - row, col
df.loc[0:2, ["city","aqi"]] 
df.columns

df.iloc[0:3, 1:5]
df.at[0, "pm25"]
df.iat[0, 5]
# df.iloc[0:2, ["city","aqi"]]

# pm25 = df["pm25"] #view
pm25 = df["city"].copy() #copy
pm25[0] = "Mumbai"

pm25
```

```python
df
```

```python
# Filtering of Data

df[ (df["aqi"] > 100)]
df[ (df["aqi"] > 100) & (df["temperature"] > 30)]

aqi_data = df[(df["aqi"] > 100) & (df["temperature"] > 30)][["city", "aqi"]]
aqi_data
aqi_data.iloc[0]
aqi_data.loc[6]
```

```python
# Query - copy
aqi_value = 50
df.query("aqi < 100 & temperature > 30")[["city", "aqi"]]

df.query("@aqi_value < 100 & temperature > 30")[["city", "aqi"]]

aqi_data = df.query("@aqi_value < 100 & temperature > 30")[["city", "aqi"]].copy()
```

```python
aqi_data
```

```python
# Cleaning Data - missing values

df = pd.read_csv("raw_data.csv")

df.isnull()
df.isna()
df.isnull().sum()

df.dropna(axis = 1) # columns drop
df.dropna() # rows drop

df.fillna(0)

cleaned_data = df.copy()
cleaned_data["age"] = cleaned_data["age"].fillna(cleaned_data["age"].mean())

cleaned_data
```

```python
df.ffill()
df.bfill()
```

```python
# Handle Duplicates

df.duplicated()
df["country"].duplicated()
df[["country", "gender"]].duplicated()

df2 = df.copy()
df.drop_duplicates(inplace=True)
```

```python
# Data Types
df.dtypes
df2 = df.copy()

df2 = df2.fillna(0)
df2["age"] = df2["age"].astype("int64").copy()
df2.dtypes
```
