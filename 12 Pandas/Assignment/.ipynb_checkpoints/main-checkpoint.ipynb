```python
import pandas as pd
```

```python
#Q1.Start by loading the IRIS dataset.
df = pd.read_csv("IRIS.csv")
```

```python
# A.Display the following information: First 10 rows, Shape & Data types, Summary statistics(mean, std, min, max)

print(f"First 10 rows : \n {df.head(10)}")

print("\n")

print(f" Shape : {df.shape} \n Data types : \n {df.dtypes}")

print("\n")

# print(f" Mean : \n {df[["sepal_length", "sepal_width", "petal_length", "petal_width"]].mean()}")
# print("\n")
# print(f" Std : \n {df[["sepal_length", "sepal_width", "petal_length", "petal_width"]].std()}")
# print("\n")
# print(f" Min : \n {df[["sepal_length", "sepal_width", "petal_length", "petal_width"]].min()}")
# print("\n")
# print(f" Max : \n {df[["sepal_length", "sepal_width", "petal_length", "petal_width"]].max()}")
print(df.describe())
```

```python
# B.Select only those rows where: petal_length > 4.5, species = "Iris-virginica"

filtered_df = df[(df.iloc[:,2] > 4.5) & (df['species'] == "Iris-virginica")]
print(filtered_df)
count = len(filtered_df)
print(f"Row Count : {count}")
```

```python
# C.Group by species and compute: Average sepal_length, Maximum petal_width, Standard deviation of sepal_width

#print(df.groupby("species")[["sepal_length","petal_width","sepal_width"]].agg(["mean", "max", "std"]))

result = df.groupby("species").agg({
    "sepal_length": "mean",
    "petal_width": "max",
    "sepal_width": "std"
})

# Optional: Rename columns for better readability
result.columns = ['Avg_Sepal_Length', 'Max_Petal_Width', 'Std_Sepal_Width']
print(result)
```

```python
# D.Createa new column: "petal_ratio"= petal_length/petal_width : For each species,find the average petal_ratio

df = df.assign(petal_ratio = df["petal_length"]/df["petal_width"])
df.groupby("species")["petal_ratio"].mean()
```

```python
# Q2.Start by loading the TITANIC dataset.
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
# A. Display only columns : Name, Sex, Age, Fare, Survived
df[["Name", "Sex", "Age", "Fare", "Survived"]]
```

```python
# B. Select passengers who: Are female, Fare > 30
filtered_df = df[(df["Sex"] == "female") & (df["Fare"] > 30)]
filtered_df
```

```python
# C. Group by Pclass and compute: Survival rate, Average Fare, Average Age
result = df.groupby("Pclass")[["Survived", "Fare", "Age"]].mean()
result.columns = ["Survival_rate", "Average_Fare", "Average_Age"]
print(result)
```
