# Python Plotting: 1D List to 2D Grid Logic

When plotting a flat list of data (like `temperatures`) onto a grid of subplots (like `2x2`), we need a way to calculate which list item corresponds to which grid cell.

### 1. The "Magic Formula" Explained

Formula: `index = i * 2 + j`

Think of the list `temperatures` as a single line of people waiting to get into a theater. You want to seat them in a theater that has **2 seats per row**.

* **`i` (The Row):** Which row you are currently filling.
* **`2` (The Width):** The total number of columns (width of the grid).
* **`j` (The Column):** The current position in that row.

**The Logic:**
> "To find the correct index, skip all the **full rows** before this one (`i * width`), and then add the **current column position** (`+ j`)."

---

### 2. Execution Trace Table

Here is how the variables change during every step of the loop:

| Iteration | `i` (Row) | `j` (Col) | Formula (`i * 2 + j`) | Resulting Index | City Selected |
| :--- | :---: | :---: | :--- | :---: | :--- |
| **1st** | `0` | `0` | `0 * 2 + 0` | **0** | New York |
| **2nd** | `0` | `1` | `0 * 2 + 1` | **1** | London |
| **3rd** | `1` | `0` | `1 * 2 + 0` | **2** | Delhi |
| **4th** | `1` | `1` | `1 * 2 + 1` | **3** | Tokyo |

---

### 3. The Universal Formula

If you have a grid with `C` columns (width), the formula to convert grid coordinates `(row, col)` to a flat list index is:

$$index = (Current Row \times Total Columns) + Current Column$$

**In Python:**
```python
index = row * width + col

````python
# Task
days = ["Mon", "Tue", "Wed", "Thu", "Fri"]

cities = ["New York", "London", "Delhi", "Tokyo"]

temperatures = [
    [22, 23, 21, 24, 25], # New York
    [18, 19, 17, 20, 21], # London
    [30, 32, 31, 33, 34], # Delhi
    [25, 26, 24, 27, 28]  # Tokyo
]
````

````python
# Using Formula
fig, ax = plt.subplots(2, 2)

for i in range(2): # rows (i) - 0,1
    for j in range(2): # cols (j) - 0,1
        row_index = i * 2 + j  # Calculate which row of temperatures to use
        ax[i][j].plot(days, temperatures[row_index])
        ax[i][j].set_title(cities[row_index])

fig.tight_layout()
plt.show()
````

````python
# Using city_no = 0
fig, ax = plt.subplots(2, 2)

city_no = 0
for i in range(2): # rows (i) - 0,1
    for j in range(2): # cols (j) - 0,1
        ax[i][j].plot(days, temperatures[city_no], marker="o")
        ax[i][j].set_title(cities[city_no])
        ax[i][j].grid(True)
        city_no += 1

fig.suptitle("Temperature in Cities over the week")
fig.supylabel("Temperature")
fig.supxlabel("Days")

fig.tight_layout()
plt.show()
````

### The `.flatten()` Method
Instead of using nested loops (rows and columns), we can flatten the grid.

1.  **`ax.flatten()`**: Converts the 2x2 matrix of plots into a straight line (1D array) of 4 plots.
2.  **Syncs with Data**: Now the plots `[0, 1, 2, 3]` align perfectly with our data lists `[0, 1, 2, 3]`.
3.  **`enumerate()`**: Gives us the index `i` (to grab data) and the plot object `ax` (to draw on) simultaneously.

**Benefit:** No more math formulas like `i*2+j`!

````python
# More Simple Apporach
fig, ax = plt.subplots(2, 2)

ax = ax.flatten()
for i, ax in enumerate(ax): # idx, val at idx
    ax.plot(days, temperatures[i], marker="o")
    ax.set_title(cities[i])
    ax.grid(True)

fig.suptitle("Temperature in Cities over the week")
fig.supylabel("Temperature")
fig.supxlabel("Days")

fig.tight_layout()
plt.show()
````
